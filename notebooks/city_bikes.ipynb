# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests
import json
import pandas as pd
import csv
```

```python
# MAKE THE API CALL TO GET ALL THE NETWORKS
url = 'https://api.citybik.es/v2/networks'
response = requests.get(url)

networks_data = response.json()
bike_networks_df = pd.json_normalize(networks_data['networks'])

# SAVE RESULTS TO CSV FILE
bike_networks_df.to_csv("../csv_folder/df_networks.csv")

```

```python
# CREATE AN EMPTY PYTHON LIST
results = []

# CALL CITYBIK API
url = "https://api.citybik.es/v2/networks/bixi-toronto"

response = requests.get(url)

if response.status_code == 200:
    results.append(response.json())
else:
    print(f"Request failed with status code {response.status_code}")
    print(f"Response content: {response.text}")

# SAVE RESULTS TO JSON FOLDER
with open("../json_folder/citybik.json", "w") as outfile:
    json.dump(results, outfile, indent=4)

print("Done. Total responses saved:", len(results))
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
# LOAD DATA FROM JSON
city_bikes = json.load(open("../json_folder/citybik.json"))

# PARSE THROUGH JSON
city_bikes_df =[]
for entry in city_bikes[0]['network']['stations']:
    city_bikes_df.append((entry['name'], entry['latitude'], entry['longitude'],entry['timestamp'],entry['free_bikes'],entry['empty_slots'], entry['extra']['renting'],entry['extra']['returning'],entry['extra']['ebikes'])) 





```

Put your parsed results into a DataFrame.

```python


city_bikes_df = pd.DataFrame(city_bikes_df, columns=['name','latitude','longitude','timestamp','free_bikes','empty_slots','renting','returning','ebikes'] )

# WRITE DATAFRAME TO CSV
city_bikes_df.to_csv("../csv_folder/city_bikes_df.csv")


city_bikes_df
```
